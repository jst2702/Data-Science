```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('ErdosNetworkExcel.txt', delimiter='\t', header = None, names = ['Source','Target','Third','Fourth'])
dataset = dataset.drop(['Third','Fourth'], axis = 1)
dataset = dataset.dropna(how='all')
#dataset = dataset.replace(np.nan,'',regex = True)
dataset
```

```python
df = dataset.copy()
df.Source = df.Source.fillna(method='ffill')
df.Source = [''.join([i for i in row if not i.isdigit()])[:-3].strip() for row in df.Source]
df
```

```python
df2 = dataset.copy().replace(np.nan,'',regex = True)
df2
coauthors = [row for row in df2.Source if len(row) > 0]
coauthors
publications = [row[-4:] for row in coauthors]
publications
nums = [int(''.join([i for i in row if i.isdigit()])) for row in publications]
publications = []
for i in nums:
    if (i < 1000):
        publications.append(i)
    else:
        publications.append(1)
coauthors = [''.join([i for i in row if not i.isdigit()])[:-3].strip() for row in coauthors]
Publishers = pd.DataFrame()
Publishers['Authors'] = coauthors
Publishers['NumPublications'] = publications
```

```python
df3 = df.copy()
df3 = df3.dropna()
df3 = df3.loc[df3['Target'].isin(df3['Source'])]
df3['Weight'] = [1 for row in df3.iterrows()]
df3.replace(',','',regex=True,inplace=True)
df3.to_csv('Edges.csv',index = False)
```

```python
groupy = df3.groupby(['Source']).sum()
l = []
for i in Publishers.Authors:
    if (i in groupy.index):
        l.append(groupy.loc[groupy.index == i].Weight[0])
    else:
        l.append(1)
```

```python
Publishers['NumCoAuthors'] = l
Publishers.replace(',','',regex=True,inplace=True)
Publishers.to_csv('Weights.csv', index = False)
```
